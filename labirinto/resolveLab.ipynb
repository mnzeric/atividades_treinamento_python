```python
import numpy as np

#lendo arquivo !!!MUDAR AQUI O NOME DO ARQUIVO!!!
txt = open('labirinto_6.txt', 'r')
conteudo = (txt.readlines())
conteudo

# retirando \n
conteudo = np.array(conteudo)
conteudo = list(map(lambda x: x.replace('\n',''), conteudo))
conteudo


# retirando as duas primeiras linha, assim o codigo le so o labirinto
info = []
info.append(conteudo.pop(0))
info.append(conteudo.pop(0))

nLC = info[1].split()
nLinhas = int(nLC[0])
nColunas = int(nLC[1])

#quebrando o tamanho do valor do indice em varios indices
lista = []
for i, j in enumerate(conteudo):
    lista.append(list(conteudo[i]))
    
#armazena em listas os pontos de entrada e saída
start, finish = [], []
for l,k in enumerate(conteudo):
    for c,n in enumerate(conteudo[l]):
        if n=='S':
            start.append(l)
            start.append(c)
            
for l,k in enumerate(conteudo):
    for c,n in enumerate(conteudo[l]):
        if n=='F':
            finish.append(l)
            finish.append(c)

            
#função
def resolveLab(lab, inicio, fim, linhas, colunas):
    #inserir aqui os pontos de entrada
    rotas = []
    caminhos = {}
    path = []
    boneco = []
    saida = [str(fim[0]),str(fim[1])]
    rotas.append([str(inicio[0]),str(inicio[1])])
    
    boneco = inicio
        
    for boneco in rotas:
        cima = int(boneco[0]) - 1
        baixo = int(boneco[0]) + 1
        esq = int(boneco[1]) - 1
        dir = int(boneco[1]) + 1
                
        if (cima > -1 and cima < linhas):
            if (lab[cima][int(boneco[1])] == '.' or lab[cima][int(boneco[1])] == 'F'):
                rotas.append([str(cima),str(boneco[1])])
                caminhos[(int(cima),int(boneco[1]))] = [int(boneco[0]),int(boneco[1])]
                if lab[cima][int(boneco[1])] != 'F':
                    lab[cima][int(boneco[1])] = 'X'
                
        if (baixo > -1 and baixo < linhas):
            if (lab[baixo][int(boneco[1])] == '.' or lab[baixo][int(boneco[1])] == 'F'):
                rotas.append([str(baixo),str(boneco[1])])
                caminhos[(int(baixo),int(boneco[1]))] = [int(boneco[0]),int(boneco[1])]
                if lab[baixo][int(boneco[1])] != 'F':
                    lab[baixo][int(boneco[1])] = 'X'
                        
        if (esq > -1 and esq < colunas):
            if (lab[int(boneco[0])][esq] == '.' or lab[int(boneco[0])][esq] == 'F'):
                rotas.append([str(boneco[0]),str(esq)])
                caminhos[(int(boneco[0]),int(esq))] = [int(boneco[0]),int(boneco[1])]
                if lab[int(boneco[0])][esq] != 'F':
                    lab[int(boneco[0])][esq] = 'X'
                        
        if (dir > -1 and dir < colunas):
            if (lab[int(boneco[0])][dir] == '.' or lab[int(boneco[0])][dir] == 'F'):
                rotas.append([str(boneco[0]),str(dir)])
                caminhos[(int(boneco[0]),int(dir))] = [int(boneco[0]),int(boneco[1])]
                if lab[int(boneco[0])][dir] != 'F':
                    lab[int(boneco[0])][dir] = 'X'
    
    if saida in rotas:
        print('SAÍDA ENCONTRADA')
        
        chave = tuple(fim)
    
        while chave != tuple(inicio):
            chave = tuple(caminhos[chave])
            if chave != tuple(inicio):
                path.append(chave)
        for ponto in path:
            lab[ponto[0]][ponto[1]] = '@'
        
        labResolvido = np.array(lab)
        print (labResolvido.reshape(nLinhas,nColunas))
    
    else:
        print('SEM SAÍDA!!')
        labResolvido = np.array(lab)
        print (labResolvido.reshape(nLinhas,nColunas))
     
#chamando a função
resolveLab(lista, start, finish, nLinhas, nColunas)
```
