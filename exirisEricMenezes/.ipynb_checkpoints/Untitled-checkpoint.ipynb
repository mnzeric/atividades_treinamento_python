```python
arqSeparado = []

#transformando o csv em uma lista    
arqBruto = open("iris.csv","r")
linhas = arqBruto.read().split("\n")
for linha in linhas:
    if linha != "":
        linha = linha.split(",")
        arqSeparado.append(linha)

def mediaGeral():
    col1,col2,col3,col4 = [],[],[],[]
    for item in arqSeparado:
        if (arqSeparado.index(item) != 0):
            col1.append(float(item[0]))
            col2.append(float(item[1]))
            col3.append(float(item[2]))
            col4.append(float(item[3]))
    m1 = sum(col1) / len(col1)
    m2 = sum(col2) / len(col2)
    m3 = sum(col3) / len(col3)
    m4 = sum(col4) / len(col4)
    
    return m1,m2,m3,m4

def mediaSetosa():
    col1,col2,col3,col4,medias = [],[],[],[],[]
    for item in arqSeparado:
        if (arqSeparado.index(item) != 0 and item[4] == '"Setosa"'):
            col1.append(float(item[0]))
            col2.append(float(item[1]))
            col3.append(float(item[2]))
            col4.append(float(item[3]))
    m1 = sum(col1) / len(col1)
    m2 = sum(col2) / len(col2)
    m3 = sum(col3) / len(col3)
    m4 = sum(col4) / len(col4)
    medias.append(str(m1))
    medias.append(str(m2))
    medias.append(str(m3))
    medias.append(str(m4))
    
    arqf = open("Setosa.csv","w")
    arqf.write(",".join(arqSeparado[0]))
    arqf.write("\n")
    arqf.write(",".join(medias))
    arqf.close()

def mediaVersicolor():
    col1,col2,col3,col4,medias = [],[],[],[],[]
    for item in arqSeparado:
        if (arqSeparado.index(item) != 0 and item[4] == '"Versicolor"'):
            col1.append(float(item[0]))
            col2.append(float(item[1]))
            col3.append(float(item[2]))
            col4.append(float(item[3]))
    m1 = sum(col1) / len(col1)
    m2 = sum(col2) / len(col2)
    m3 = sum(col3) / len(col3)
    m4 = sum(col4) / len(col4)
    medias.append(str(m1))
    medias.append(str(m2))
    medias.append(str(m3))
    medias.append(str(m4))
    
    arqf = open("Versicolor.csv","w")
    arqf.write(",".join(arqSeparado[0]))
    arqf.write("\n")
    arqf.write(",".join(medias))
    arqf.close()
    
def mediaVirginica():
    col1,col2,col3,col4,medias = [],[],[],[],[]
    for item in arqSeparado:
        if (arqSeparado.index(item) != 0 and item[4] == '"Virginica"'):
            col1.append(float(item[0]))
            col2.append(float(item[1]))
            col3.append(float(item[2]))
            col4.append(float(item[3]))
    m1 = sum(col1) / len(col1)
    m2 = sum(col2) / len(col2)
    m3 = sum(col3) / len(col3)
    m4 = sum(col4) / len(col4)
    medias.append(str(m1))
    medias.append(str(m2))
    medias.append(str(m3))
    medias.append(str(m4))
    
    arqf = open("Virginica.csv","w")
    arqf.write(",".join(arqSeparado[0]))
    arqf.write("\n")
    arqf.write(",".join(medias))
    arqf.close()
    


mediaSetosa()
mediaVersicolor()
mediaVirginica()
```
